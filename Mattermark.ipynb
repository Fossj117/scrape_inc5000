### 1. What Data Can Be Collected?

The most obvious, useful, and straightforward data to collect on the Inc. 5000 website is all the data displayed on each company when a user clicks through to a page like [this](http://www.inc.com/profile/fuhu). 

This includes the following: 

* **Name**
* **Rank** (2014)
* **Location** (and rank w/in location)
    * Metro area (e.g. "Los Angeles")
    * Specific city (e.g. "El Segundo")
* **Industry** (and rank w/in industry)
* **Business model description** ("The creator of an Android tablet for kids...")
* **Number of employees**
* **Number of employees 4 yrs ago** (i.e. employees added in last three years)
* **Revenues** (2014)
* **Revenues 4 yrs ago**
* **Twitter handle**
* **Website URL**
* **Date Founded**

One approach to collecting this data would be to load the main listing page, and then follow through all the company links, and then screen scrape the data from the display. However, after playing with the site a bit, I realized there is a much easier way. 

Essentially, when the site goes to load a page like [this](http://www.inc.com/profile/fuhu) one (for Fuhu), it gets all the necessary data to dynamically generate the HTML by pining a `GET` request to the `'http://www.inc.com/rest/inc5000company/` endpoint, and passing the `company_id`. This returns a JSON object like this: 

```
{"app_revenues_fouryearsago": "123000", "city_display_name": "Los Angeles", "app_employ_fouryearsago": "49", "rank": "1", "years": [{"ify_industry_rank": "1", "ify_metro_rank": "1", "ify_state_rank": "1", "ify_year": "2014", "ify_rank": "1"}, {"ify_industry_rank": null, "ify_metro_rank": null, "ify_state_rank": null, "ify_year": "2013", "ify_rank": "1"}], "app_revenues_lastyear": "195640000", "ifc_filelocation": "fuhu", "id": "22890", "app_employ_lastyear": "227", "ifc_twitter_handle": "NabiTablet", "industry_id": "4", "ifc_business_model": "The creator of an Android tablet for kids and an Adobe Air application that allows children to access the Internet in a parent-controlled environment.", "ifc_url": "http://www.fuhu.com", "next_id": "25747", "ifc_state": "CA", "metro_id": "2", "industry_rank": "1", "ifc_city": "El Segundo", "ifi_industry": "Consumer Products & Services", "ifc_company": "Fuhu", "state_rank": "1", "metro_rank": "1", "current_industry_rank": "1", "ifc_founded": "2008-00-00"}
```

This JSON contains all the important data about the company. It also includes a `next_id` field which points to the next company in the list (so that the browser knows what to grab next if the user scrolls down). As such, it's a straightforward matter to use python to mimic this behavior and store the data as JSON in a text file. I have done this in the `scrape_inc5000.py` file, which can be found in the [github for this project](https://github.com/Fossj117/scrape_inc5000). With this explanation in mind, hopefully everything in the script should be self-explanatory. Running the script will repeatedly query the internal API endpoint with the `requests` package and dump the results into a file `results.json` (which can then be read with pandas or whatever). 

I ran the script using [domino](http://www.dominoup.com/) for convenience (it takes 10-15 min) and have included the resulting raw `results.json` file in the github.

### 3. Matching to Crunchbase

The best way to match these up to crunchbase profiles would depend a bit on exactly what data we had from crunchbase. I that we probably have at least the follwing from Crunchbase: 

* Company name
* HQ location
* Description
* Website
* Date founded

We have similar fields in the data collected from the Inc 5000 listings. In considering which of these to match, would be good to consider which fields are most robust (i.e. stable) and least likely to be missing. If `website` tends to reliably be present in both datasets, that could be a good candidate. Some sort of fuzzy matching on the location data could work as well (see musings below). 

A method that would definitely work (but could be somewhat messy) would be to use something like [Levenshtein Distance](http://en.wikipedia.org/wiki/Levenshtein_distance) to do fuzzy matching on lowercased company names. The idea would be: 

```
for each listing in the Inc 5000 dataset: 
        for each crunchbase listing:
            compute levenshtein distance and store best match (or top 5)
```
This strategy could be constrained using the date founded or location fields if compute time were an issue. A nice feature of the strategy is also that listings *without* a really strong match (i.e. nothing within some lev. distance threshold) can be returned as points for manual inspection. We could also extend the method by incorporating website and/or location into our similarity score, which would likely make the method more robust.
